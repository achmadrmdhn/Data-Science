```python
# 1Ô∏è‚É£ Membuat DataFrame Contoh
data = {
    "nama": ["Rifai", "Dika", "Siti", "Budi", "Dika", "Ani", "Rifai"],
    "usia": [20, None, 21, 22, 22, None, 20],  # Ada nilai NaN (None)
    "kota": ["Jakarta", "Bandung", None, "Surabaya", "Bandung", "Jakarta", "Jakarta"],  # Ada nilai kosong
    "tanggal_lahir": ["2004-05-12", "2003-08-21", "2002-12-11", "2002-01-15",
                      "2003-08-21", "2005-03-27", "2004-05-12"]  # Tanggal dalam string
}
```

```python
df = pd.DataFrame(data)
print("üìå Data Awal:")
df
```

```python
# 2Ô∏è‚É£ Menghapus duplikasi
df = df.drop_duplicates()
```

```python
# 3Ô∏è‚É£ Menangani missing values
df["usia"] = df["usia"].fillna(df["usia"].mean())  # Isi NaN di usia dengan rata-rata usia
df = df.dropna(subset=["kota"])  # Hapus baris dengan kota yang kosong
```

```python
# 4Ô∏è‚É£ Mengubah tipe data tanggal
df["tanggal_lahir"] = pd.to_datetime(df["tanggal_lahir"])
```

```python
# 5Ô∏è‚É£ Menambah kolom tahun lahir
df["tahun_lahir"] = df["tanggal_lahir"].dt.year
```

```python
# 6Ô∏è‚É£ Standarisasi format teks
df["nama"] = df["nama"].str.title()
```

```python
print("\n‚úÖ Data Setelah Wrangling:")
df
```

Mengisi Missing Values

```python
df["Bulan"] = pd.to_datetime(df["Tanggal"]).dt.month

# Mengisi missing values berdasarkan rata-rata per bulan
for col in ["Suhu_RataRata", "Total_Curah_Hujan", "Total_Salju", "Salju_DiTanah"]:
    df[col] = df.groupby("Bulan")[col].transform(lambda x: x.fillna(x.mean()))

df.head()
```

buat diagram (plot) yang menunjukkan

```python
# Rata-rata Suhu Harian selama Setahun
plt.figure(figsize=(12, 5))
plt.plot(df["Tanggal"], df["Suhu_RataRata"], label="Suhu Rata-Rata (¬∞C)", color="blue")
plt.xlabel("Tanggal")
plt.ylabel("Suhu (¬∞C)")
plt.title("Rata-Rata Suhu Harian Selama Setahun")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ketebalan Salju Selama Setahun
plt.figure(figsize=(12, 5))
plt.plot(df["Tanggal"], df["Salju_DiTanah"], label="Ketebalan Salju (cm)", color="green")
plt.xlabel("Tanggal")
plt.ylabel("Ketebalan Salju (cm)")
plt.title("Rata-Rata Ketebalan Salju Selama Setahun")
plt.legend()
plt.grid(True)
plt.show()
```
