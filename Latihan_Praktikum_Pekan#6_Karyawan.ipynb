```python
# Membaca atau menghubungkan direktori folder dataset yang ada di google drive
path = "/content/gdrive/MyDrive/Datasets/DS/data/"
```

```python
# Membaca dan menampilkan file
df1 = pd.read_csv(path + 'karyawan-penilaian2018.csv')
df2 = pd.read_csv(path + 'karyawan-penilaian2019.csv')
df3 = pd.read_csv(path + 'karyawan1.csv')
df4 = pd.read_csv(path + 'karyawan2.csv')
```

```python
# Standarisasi nama kolom
# df1 = df1.rename(columns={"ID": "id_karyawan", "Nama Pegawai": "nama_pegawai", "Penilaian 2018": "penilaian_2018"})
df1 = df1.rename(columns={"id_karyawan": "ID", "nama_pegawai": "Nama Pegawai", "penilaian_2018": "Penilaian 2018"})
# df2 = df2.rename(columns={"ID": "id_karyawan", "Nama Pegawai": "nama_pegawai", "Penilaian 2019": "penilaian_2019"})
df2 = df2.rename(columns={"id_karyawan": "ID", "nama_pegawai": "Nama Pegawai"})
```

```python
# df1
# df2
```

```python
# Gabung penilaian
# join_penilaian = pd.merge(df1, df2, on=["id_karyawan", "nama_pegawai"], how="outer")
join_penilaian = pd.merge(df1, df2, on=["ID", "Nama Pegawai"], how="outer")
join_penilaian
```

1. siapa saja karyawan yang performanya naik dari 2018 ke 2019

```python
performa_naik = join_penilaian[
    (join_penilaian["Penilaian 2018"].notna()) &
    (join_penilaian["Penilaian 2019"].notna()) &
    (join_penilaian["Penilaian 2019"] > join_penilaian["Penilaian 2018"])
]
performa_naik
```

2. Siapa saja karyawan baru di 2019

```python
karyawan_baru_2019 = join_penilaian[join_penilaian["Penilaian 2018"].isna()]
karyawan_baru_2019
```

3. Siapa saja yang keluar di 2019

```python
karyawan_keluar_2019 = join_penilaian[join_penilaian["Penilaian 2019"].isna()]
karyawan_keluar_2019
```

4. Tabel gabungannya di simpan sebagai file evaluasi-karyawan.csv

```python
join_penilaian.to_csv("evaluasi-karyawan.csv", index=False)
```

5. Gabungkan kedua file tersebut
6. Simpan sebagai data-karyawan.csv

```python
data_karyawan = pd.concat([df3, df4], ignore_index=True)
data_karyawan.to_csv("data-karyawan.csv", index=False)
```
