Berapa rata-rata harga rumah yang ada di data tersebut?

```python
average_price = df["price"].mean()
average_price
```

Rumah mana yang paling tua? Berapa harganya? Apakah mahal atau murah?

```python
oldest_house = df[df["yr_built"] == df["yr_built"].min()]
oldest_house_price = oldest_house["price"].values[0]
price_category = "mahal" if oldest_house_price > average_price else "murah"

print( oldest_house["yr_built"].values[0])
print(oldest_house_price)
print(price_category)
```
