```python
# Mengubah tipe data kolom tanggal lahir
df['Tanggal Lahir'] = pd.to_datetime(df['Tanggal Lahir'], format='%d-%m-%Y', errors='coerce')
```

```python
# Menormalisasi data jenis kelamin
gender_map = {
    'L': 'Laki-Laki',
    'Pria': 'Laki-Laki',
    'P': 'Perempuan',
    'Wanita': 'Perempuan',
    'W': 'Perempuan'
}
df['Jenis Kelamin'] = df['Jenis Kelamin'].map(gender_map)
```

```python
# Menormalisasi data jam kerja
def normalize_time(value):
    if pd.isna(value):
        return np.nan

    value_str = str(value).strip().lower()

    # Menghapus kata 'wib' dan lainnya
    value_str = value_str.replace('wib', '').replace('.', ':').strip()

    # Menangani format AM/PM
    if 'am' in value_str or 'pm' in value_str:
        try:
            return pd.to_datetime(value_str, format='%I %p').strftime('%H:%M')
        except:
            return np.nan

    # Menangani jika jamnya hanya dua digit (contoh: 88 -> 08:08)
    if value_str.isdigit():
        if len(value_str) == 4:
            return f"{value_str[:2]}:{value_str[2:]}"
        elif len(value_str) == 2:
            return f"0{value_str[0]}:0{value_str[1]}"
        elif len(value_str) == 1:
            return f"0{value_str}:00"

    # parse sebagai HH:MM
    try:
        return pd.to_datetime(value_str, format='%H:%M').strftime('%H:%M')
    except:
        pass

    # parse sebagai HH
    try:
        return pd.to_datetime(value_str, format='%H').strftime('%H:%M')
    except:
        return np.nan
df['Jam Kerja'] = df['Jam Kerja'].apply(normalize_time)
```

```python
# Menormalisasi data berat badan
df['Berat Badan'] = (
    df['Berat Badan']
    .astype(str)
    .str.replace(r'[^\d,\.]', '', regex=True)
    .str.replace(',', '.')
)

df['Berat Badan'] = pd.to_numeric(df['Berat Badan'], errors='coerce')

df['Berat Badan'] = df['Berat Badan'].apply(
    lambda x: x/10 if x > 200 else x
)
```

```python
# Mengubah data No HP menjadi numerik
def normalize_phone(number):
   try:
       clean_number = str(int(float(number)))

       if clean_number.startswith('0'):
           return '+62' + clean_number[1:]
       else:
           return '+62' + clean_number

   except:
       return np.nan

df['No HP'] = df['No HP'].apply(normalize_phone)
```
